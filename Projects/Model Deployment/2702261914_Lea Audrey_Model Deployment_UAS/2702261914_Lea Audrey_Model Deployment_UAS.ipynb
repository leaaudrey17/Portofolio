## Preprocessing

```python
# Mapping FCVC (frekuensi konsumsi sayur)
def map_fcvc(x):
    if x < 2:
        return 'Jarang'
    elif x == 2:
        return 'Sedang'
    else:
        return 'Sering'

# Mapping NCP (jumlah makan per hari)-> makan sehari normalnya 3x sehari
def map_ncp(x):
    if x < 3:
        return 'Kurang'
    elif x == 3:
        return 'Cukup'
    else:
        return 'Berlebih'

# Mapping CH2O (konsumsi air)
def map_ch2o(x):
    if x <= 1.5:
        return 'Kurang'
    elif x <= 2.5:
        return 'Cukup'
    else:
        return 'Banyak'

# Mapping FAF (Frekuensi aktivitas fisik 03)
def map_faf(x):
    if x == 0:
        return 'Tidak Aktif'
    elif x <= 1:
        return 'Kurang Aktif'
    elif x <= 2:
        return 'Cukup Aktif'
    else:
        return 'Sangat Aktif'

# Mapping TUE (Penggunaan teknologi harian 03)
def map_tue(x):
    if x == 0:
        return 'Tidak Pernah'
    elif x <= 1:
        return 'Jarang'
    elif x <= 2:
        return 'Sering'
    else:
        return 'Sangat Sering'

# Terapkan mapping ke kolom
df['FCVC'] = df['FCVC'].apply(map_fcvc)
df['NCP'] = df['NCP'].apply(map_ncp)
df['CH2O'] = df['CH2O'].apply(map_ch2o)
df['FAF'] = df['FAF'].apply(map_faf)
df['TUE'] = df['TUE'].apply(map_tue)
```

## Pipeline

### Preprocessing pipelien

```python
ordinal_features = ['FCVC', 'NCP', 'CH2O', 'FAF', 'TUE']
ordinal_categories = [
    ['Jarang', 'Sedang', 'Sering'],                             # FCVC
    ['Kurang', 'Cukup', 'Berlebih'],                            # NCP
    ['Kurang', 'Cukup', 'Banyak'],                              # CH2O
    ['Tidak Aktif', 'Kurang Aktif', 'Cukup Aktif', 'Sangat Aktif'], # FAF
    ['Tidak Pernah', 'Jarang', 'Sering']                       # TUE
]

nominal_features = ['Gender', 'family_history_with_overweight', 'FAVC',
                    'CAEC', 'SMOKE', 'SCC', 'CALC', 'MTRANS']

numeric_features = ['Age', 'Height', 'Weight']

target = 'NObeyesdad'
```

```python
from sklearn.preprocessing import OrdinalEncoder, OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

# Ordinal pipeline
ordinal_pipeline = Pipeline([
    ('ordinal_enc', OrdinalEncoder(categories=ordinal_categories))
])

# Nominal pipeline
nominal_pipeline = Pipeline([
    ('onehot_enc', OneHotEncoder(handle_unknown='ignore'))
])

# Numeric pipeline
numeric_pipeline = Pipeline([
    ('scaler', StandardScaler())
])

# Full preprocessing
preprocessor = ColumnTransformer([
    ('ord', ordinal_pipeline, ordinal_features),
    ('nom', nominal_pipeline, nominal_features),
    ('num', numeric_pipeline, numeric_features)
])
```

```python
target_order = [
    'Insufficient_Weight',
    'Normal_Weight',
    'Overweight_Level_I',
    'Overweight_Level_II',
    'Obesity_Type_I',
    'Obesity_Type_II',
    'Obesity_Type_III'
]

# OrdinalEncoder
ordinal_enc = OrdinalEncoder(categories=[target_order])
df[target] = ordinal_enc.fit_transform(df[[target]]).astype(int)
```

# Model

### Split data

```python
# Split the data into features (X) and target (y)
X = df.drop(columns=target)  # Features
y = df[target]  # Target column
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=42)
```

# Random Forest Classifier

```python
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline

# Preprocessor dan classifier
pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('classifier', RandomForestClassifier(random_state=42))
])

# Hyperparamater
param_grid = {
    'classifier__n_estimators': [100, 200, 300],                # Number of trees in the forest
    'classifier__max_depth': [10, 20, 30],                        # Maximum depth of each tree
    'classifier__min_samples_split': [2, 5, 10],                  # Minimum number of samples required to split a node
    'classifier__min_samples_leaf': [1, 2, 4]                    # Minimum number of samples required at a leaf node
}

# Tuning
grid_search = GridSearchCV(estimator=pipeline, param_grid=param_grid, cv=5, scoring='f1_weighted', n_jobs=-1)
grid_search.fit(X_train, y_train)

## Evaluasi

# Best parameters from GridSearchCV
print("Best Parameters from Grid Search:")
print(grid_search.best_params_)

# Predict using the best estimator
y_pred_tuned = grid_search.best_estimator_.predict(X_test)

# Confusion Matrix
print("Random Forest with GridSearchCV - Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_tuned))

## Classification Report
print("\nRandom Forest with GridSearchCV - Classification Report:")
print(classification_report(y_test, y_pred_tuned))
```

## 1. **Hasil GridSearchCV  Best Parameters**

* **`n_estimators = 200`**
  Jumlah pohon (trees) dalam forest. Nilai ini cukup besar dan bagus untuk kestabilan prediksi. Lebih banyak pohon bisa meningkatkan akurasi, tapi juga meningkatkan waktu komputasi.

* **`max_depth = 20`**
  Kedalaman maksimal tiap pohon adalah 20 level. Ini mencegah overfitting dengan membatasi seberapa dalam pohon bisa tumbuh. Nilai ini memberi keseimbangan antara kompleksitas dan generalisasi.

* **`min_samples_split = 5`**
  Sebuah node akan di-split jika minimal ada 5 sampel di dalamnya. Ini mencegah split yang terlalu dini, membantu mengurangi overfitting.

* **`min_samples_leaf = 1`**
  Jumlah minimum sampel di daun adalah 1. Artinya, model bisa membuat node akhir walau hanya ada 1 data di dalamnya  default Random Forest, tapi masih bisa bekerja dengan baik bila data besar dan bersih.

# XGBoost

```python
# XGBoost Model
import xgboost as xgb
from sklearn.preprocessing import LabelEncoder # Also import LabelEncoder
from sklearn.model_selection import train_test_split # Import train_test_split
from sklearn.model_selection import GridSearchCV # Import GridSearchCV
from sklearn.metrics import confusion_matrix, classification_report # Import metrics
from sklearn.pipeline import Pipeline # Import Pipeline
xgb_model = xgb.XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='mlogloss')

# Create a pipeline that includes preprocessing and the XGBoost model
pipeline = Pipeline([
    ('preprocessing', preprocessor),
    ('classifier', xgb_model)
])


# Define the hyperparameter grid for XGBoost
param_grid_xgb = {
    'classifier__n_estimators': [100, 200, 300],
    'classifier__max_depth': [3, 6, 10],
    'classifier__learning_rate': [0.01, 0.1, 0.3],
    'classifier__subsample': [0.7, 0.8, 1.0],
    'classifier__colsample_bytree': [0.7, 0.8, 1.0]
}

# GridSearchCV for hyperparameter tuning
grid_search_xgb = GridSearchCV(estimator=pipeline, param_grid=param_grid_xgb, cv=5, scoring='f1_weighted', n_jobs=-1)

# Fit the model with GridSearchCV
grid_search_xgb.fit(X_train, y_train)

# Best parameters from GridSearchCV
print("Best Parameters for XGBoost:")
print(grid_search_xgb.best_params_)

# Predict using the best estimator
y_pred_xgb = grid_search_xgb.best_estimator_.predict(X_test)

# Model Evaluation for XGBoost
print("XGBoost with GridSearchCV - Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_xgb))

print("\nXGBoost with GridSearchCV - Classification Report:")
print(classification_report(y_test, y_pred_xgb))
```

## 1. **Best Parameters dari GridSearchCV**

* **colsample\_bytree = 1.0**
  Artinya seluruh fitur digunakan untuk membangun setiap pohon. Jika lebih rendah (misal 0.8), hanya sebagian fitur dipakai untuk membuat tiap tree.

* **learning\_rate = 0.1**
  Merupakan **ukuran langkah** dalam mengupdate model. Nilai ini cukup umum dan seimbang antara kecepatan dan akurasi.

* **max\_depth = 6**
  Kedalaman maksimal tiap pohon decision tree. Ini membatasi kompleksitas agar tidak overfit.

* **n\_estimators = 200**
  Jumlah pohon (boosting rounds). Lebih banyak pohon bisa memperbaiki akurasi, tapi juga lebih berat dihitung.

* **subsample = 1.0**
  Artinya 100% data digunakan untuk training setiap pohon. Jika diset ke 0.8, maka hanya 80% data per pohon  sering dipakai untuk regularisasi.
